```python
import pandas as pd
import numpy as np 
```

```python
url = "./breast-cancer.csv"
data = pd.read_csv(url)
```

```python

data['diagnosis'] = data['diagnosis'].replace({'M': 1, 'B': 0})
# Lấy cột cần di chuyển
column_to_move = data['diagnosis']
data.drop(columns=["id"], inplace=True)
# Loại bỏ cột này khỏi DataFrame
data.drop(columns=['diagnosis'], inplace=True)
# Thêm cột này lại vào cuối DataFrame
data['diagnosis'] = column_to_move
data.head(20)
```

```python
from sklearn.utils import resample
# Tách dữ liệu thành các lớp benign và malignant
benign = data[data['diagnosis'] == 0]
malignant = data[data['diagnosis'] == 1]
# Tạo hàm augmentation
def augment_data(data, n_samples):
    augmented_data = []
    for _ in range(n_samples):
        # Lấy ngẫu nhiên một mẫu từ dữ liệu hiện có
        sample = data.sample()
        # Tạo một phiên bản mới của mẫu bằng cách thay đổi một số thuộc tính ngẫu nhiên
        augmented_sample = sample.apply(lambda x: x * np.random.uniform(0.9, 1.1) if np.random.rand() < 0.5 else x)
        augmented_data.append(augmented_sample)
    return pd.concat(augmented_data)
# Số lượng mẫu mới bạn muốn tạo cho mỗi lớp
n_samples = 1000
# Sinh dữ liệu mới cho cả hai lớp
augmented_benign = augment_data(benign, n_samples)
augmented_malignant = augment_data(malignant, n_samples)
# Kết hợp dữ liệu mới với dữ liệu gốc
augmented_data = pd.concat([data, augmented_benign, augmented_malignant])
# In số lượng mẫu mới của mỗi lớp
print("Số lượng mẫu mới benign:", len(augmented_benign))
print("Số lượng mẫu mới malignant:", len(augmented_malignant))

# In số lượng mẫu của mỗi lớp sau khi augmentation
print("Số lượng mẫu sau khi augmentation:", len(augmented_data))
```
